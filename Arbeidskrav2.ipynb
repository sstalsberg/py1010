```python
data = {"Norge": ["Oslo", 0.634],
        "England": ["London", 8.982],
        "Frankrike": ["Paris", 2.161],
        "Italia": ["Roma", 2.873]
        }

land = input("Skriv inn navnet på et land:")
hovedstad = data[land][0]
innbyggere = data[land][1]

print(hovedstad, "er hovedstaden i", land, "og det er", innbyggere, "mill. innbyggere i", hovedstad)

nyland = input("Skriv inn navnet på ett nytt land:")
nyhovedstad = input("Skriv inn hovedstaden på det nye landet:")
nyinnbyggere = float(input("Skriv inn antall millioner innbyggere i hovedstaden i det nye landet:"))
data[nyland] = [nyhovedstad, nyinnbyggere]
print(data)
```

####Oppgave 5

Lag et program med en funksjon som tar a og b som inn-argumenter og som så
regner ut arealet og «ytre» omkrets til en figur satt sammen av en rettvinklet trekant og en halvsirkel, se figuren under. Med «ytre» omkrets menes samlet lengde av de sorte strekene. Funksjonen skal returnere arealet og «ytre» omkrets, som så skrives til skjerm med passende tekst.Screenshot 2024-12-27 at 18.24.53.png

```python
from math import sqrt
import numpy as np

a = int(input("Tast inn et heltall (a): "))
b = int(input("Tast inn et annet heltall (b): "))

sirkel_a = ((np.pi*(a*0.5))**2)/2
sirkel_o = (np.pi*a)/2
trekant_a = (a*b)/2
trekant_o = b+sqrt((a*a)+(b*b))

print("Den ytre omkretsen til figuren er: ", round(sirkel_o+trekant_o, 2))
print("Arealet til figuren er: ", round(sirkel_a+trekant_a, 2))
```

####Oppgave 6

Skriv en kode som plotter **funksjonen** 𝑓(𝑥) = −𝑥2 − 5, for x på intervallet [-10,10].

Hint: np.linspace(-10, 10, 200) gir en array med 200 punkter jevnt fordelt på intervallet [-10,10]

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)
y = -x**2-5

plt.plot(x, y)
plt.show()
```
